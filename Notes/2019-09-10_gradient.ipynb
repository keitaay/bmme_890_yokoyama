```python
import numpy as np
import pylab as plt
from pylab import meshgrid,cm,imshow

%matplotlib inline

b=[2, 10]
x=(np.arange(0,100,1))

def z(x1,x2,b):
    return (b[0]*x1 + b[1]*x2)

# grid of points
X1,X2 = meshgrid(x,x)

# evaluate function on above grid
Z=z(X1,X2,b)

# display result
im=imshow(Z,cmap='gray')
```

```python
from mpl_toolkits.mplot3d import Axes3D

%matplotlib auto

fig=plt.figure()

ax=fig.gca(projection='3d')
surf=ax.plot_surface(X1,X2,Z)
```

```python
%matplotlib inline

# the gradient function flips axes from the "usual" Python convention to first axis
Zx_2,Zx_1 = np.gradient(Z)

plt.subplot(1,2,1)
plt.imshow(Zx_1, vmin=0,vmax=10)
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(Zx_2, vmin=0,vmax=10)
plt.colorbar()
```
