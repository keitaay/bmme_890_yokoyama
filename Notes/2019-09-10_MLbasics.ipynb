## ML is good for:

* problems requiring lots of fine-tuning outputs / long list of rules

* complex problems with poor traditional solutions

* fluctuation in data

* gaining non-obvious insights about complex problems/big data

## Type of ML

* existence of human supervision (supervised, unsupervised, semisurpervised, reinforcement)

* capacity of learning incrementally on the fly (online vs batch)

* algorithm compares new data to known data points OR detect patterns in training data to build predictive model (instance-based vs model-based)

### Supervision

#### Supervised Learning

Learn from the past or learn from someone else

* classification - provide label at end of training

* regression

Dimensionality reduction simplifies data without losing too much information. It could be done, for example, by merging several correlated features into a single dimension.
(e.g. car's mileage may be correlated with age, so the dimensionality reduction algorithm will merge them into one feature to represent the car's wear and tear)

This is **feature extraction**, and it's a very important step in ML.

#### Semisupervised Learning

Labeling data is usually time-consuming and costly, and data is often unlabeled. ...but some algorithms can deal with data that's partially labeled.

(e.g. Google Photo notices the same person in multiple uploaded images; user identifies people; algorithm extends identification to others)

#### Reinforcement Learning

An agent observes the environment, selects/oerforms action, and gets a reward in return.

The algorithm must, then, learn by itself what the best strategy (policy) is that gets the most reward over time.

A policy defines what action the agent should choose in a given situation.


### Dataset organization

#### Batch learning

* system is incapable of learning incrementally; must be trained using all available data (lots of time, resources)
* incorporating new data requires new system version
* training using full dataset can take many hours

#### Online learning

* feed data into system incrementally/sequentially, individually or by small groups (mini-batches)
* fast, cheap, can learn about new data on the fly
* important parameter: speed of adapting to changing data: **learning rate**
